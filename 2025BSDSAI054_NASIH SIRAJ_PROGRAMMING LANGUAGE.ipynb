```python
#Problem 1
#part 1
def gcd(m,n):
    while n!=0:
        rem=m%n
        m=n
        n=rem
    return m
print(gcd(24,30))
print(gcd(-24,-30))
```

```python
# problem 3
# part 1
def is_power_of_2(n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if 2**i==n:
                return True
        return False
is_power_of_2(15)
```

```python
#part 2
def is_power(b,n):
    if n==0:
        return False
    elif n==1:
        return True
    else:
        for i in range(1,n):
            if b**i==n:
                return True
        return False
is_power(8,64)
```
